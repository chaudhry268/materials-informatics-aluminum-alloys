```python
import pptx
from pptx.util import Inches

# Create a presentation
ppt = pptx.Presentation()

# Define slide layout (Title and Content)
title_slide_layout = ppt.slide_layouts[0]
content_slide_layout = ppt.slide_layouts[1]

# Slide 1: Title Slide
slide = ppt.slides.add_slide(title_slide_layout)
title = slide.shapes.title
subtitle = slide.placeholders[1]
title.text = "Truck Art of Pakistan"
subtitle.text = "A Cultural Canvas on Wheels\nPresented by Waseem"

# Slide 2: Introduction
slide = ppt.slides.add_slide(content_slide_layout)
title = slide.shapes.title
content = slide.placeholders[1]
title.text = "What is Truck Art?"
content.text = "Truck Art is a vibrant, colorful form of artistic expression that transforms transport vehicles into moving masterpieces."

# Slide 3: History & Origins
slide = ppt.slides.add_slide(content_slide_layout)
title = slide.shapes.title
content = slide.placeholders[1]
title.text = "History & Origins"
content.text = "Truck Art in Pakistan began in the 1940s with vehicle decorations by craftsmen. It evolved into a cultural identity, influenced by traditional motifs and local aesthetics."

# Slide 4: Cultural Significance
slide = ppt.slides.add_slide(content_slide_layout)
title = slide.shapes.title
content = slide.placeholders[1]
title.text = "Cultural Significance"
content.text = "More than just decoration, Truck Art reflects national pride, folklore, religious themes, and even poetry, telling stories through colors and images."

# Slide 5: Artistic Elements
slide = ppt.slides.add_slide(content_slide_layout)
title = slide.shapes.title
content = slide.placeholders[1]
title.text = "Artistic Elements"
content.text = "• Bright colors & floral motifs\n• Calligraphy & poetry\n• Portraits of heroes & celebrities\n• Decorative accessories like chains, bells, and mirrors"

# Slide 6: Global Recognition
slide = ppt.slides.add_slide(content_slide_layout)
title = slide.shapes.title
content = slide.placeholders[1]
title.text = "Impact & Global Recognition"
content.text = "Pakistani Truck Art has gained international recognition, influencing fashion, furniture, and even aircraft designs worldwide."

# Slide 7: Modern Trends & Future
slide = ppt.slides.add_slide(content_slide_layout)
title = slide.shapes.title
content = slide.placeholders[1]
title.text = "Modern Trends & Future"
content.text = "Digital truck art, exhibitions, and customized art projects are taking this traditional form into the modern era."

# Save the PowerPoint file
ppt_path = "/mnt/data/Truck_Art_of_Pakistan.pptx"
ppt.save(ppt_path)
ppt_path
```
